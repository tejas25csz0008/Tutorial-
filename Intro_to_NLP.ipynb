**2.1 Lowercasing**

Lowercasing converts all text into small letters.

Why?

Because computers treat

Drone ≠ drone ≠ DRONE

So we standardize text.

```python
text = "Drone Detected Near Base"

lower_text = text.lower()

print(lower_text)
```

**2.2 Tokenization**

Tokenization = Splitting sentences into words.

**Why Is Tokenization Needed?**

Computers cannot understand full sentences directly.

They process text word by word.

So we break text into pieces first.

Examples:

"Enemy entered restricted area"

['Enemy', 'entered', 'restricted', 'area']

```python
!pip install nltk
```

**2.3 Stopword Removal**

Stopwords are frequently occurring words that do not add significant meaning.

Stopwords = Common words with little meaning.

**Examples:** is, the, at, on, and

"The enemy is near the border"

enemy near border

**Example**

**Before Removing Stopwords (Original Paragraph)**

The soldiers were patrolling the border area during the night.

They were watching all the movements near the river.

The team was fully alert and ready for any situation.

They were communicating with the base regularly.

The operation was completed successfully.


**After Removing Stopwords**

soldiers patrolling border area night

watching movements near river

team fully alert ready situation

communicating base regularly

operation completed successfully

**Even after removing stopwords, we still understand:**

Who → soldiers / team

Action → patrolling, watching, communicating

Location → border, river, base

Status → alert, successful

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

tokens = ["mission", "was", "successful", "and", "unit", "returned"]

filtered_tokens = [word for word in tokens if word not in stop_words]

print(filtered_tokens)
```

**2.5 Lemmatization**

Lemmatization also reduces words to root form
BUT returns meaningful dictionary words.

| Original | Lemma |
| -------- | ----- |
| running  | run   |
| better   | good  |
| was      | be    |


```python
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

words = ["running", "better", "was"]

lemmas = [lemmatizer.lemmatize(word) for word in words]

print(lemmas)
```

Lemmatization = Smart root conversion using vocabulary.

More accurate than stemming.

| Word     | Stemming | Lemmatization |
| -------- | -------- | ------------- |
| running  | run      | run           |
| better   | better   | good          |
| was      | wa       | be            |
| security | secur    | security      |

**SECTION 3: Text Representation**

Computers don’t understand words — they understand numbers.

So we convert text → numerical form.



**3.1 Bag of Words (BoW)**

BoW converts text into word frequency vectors.

Example Sentences

“Enemy spotted”

“Enemy moving”

Enemy, spotted, moving


| Sentence | Enemy | spotted | moving |
| -------- | ----- | ------- | ------ |
| 1        | 1     | 1       | 0      |
| 2        | 1     | 0       | 1      |


```python
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
    "Enemy spotted",
    "Enemy moving"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(sentences)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

Problem with BoW:

Common words dominate counts.

TF-IDF gives importance to rare but meaningful words.

**Example:**

1. enemy detected near border
2. enemy moving towards border
3. enemy entered border area


Bag of words

First list unique words:

enemy, detected, near, border, moving, towards, entered, area


| Sentence | enemy | border | detected | moving | entered | near | towards | area |
| -------- | ----- | ------ | -------- | ------ | ------- | ---- | ------- | ---- |
| S1       | 1     | 1      | 1        | 0      | 0       | 1    | 0       | 0    |
| S2       | 1     | 1      | 0        | 1      | 0       | 0    | 1       | 0    |
| S3       | 1     | 1      | 0        | 0      | 1       | 0    | 0       | 1    |


**Problem with BoW**

Look at frequencies:

enemy → appears in all sentences

border → appears in all sentences

So BoW thinks:

enemy, border = most important words

But in reality:

detected

moving

entered

tell us what actually happened.

So BoW over values common words.

**TF-IDF Idea**

TF-IDF reduces weight of common words and increases weight of rare words.

Concept:

Words in every sentence → Low importance

Words in few sentences → High importance

| Word     | Frequency | Importance |
| -------- | --------- | ---------- |
| enemy    | High      | Low        |
| border   | High      | Low        |
| detected | Low       | High       |
| moving   | Low       | High       |
| entered  | Low       | High       |


**Practical Analogy**

BoW = Counting how many times soldiers speak

TF-IDF = Measuring who gives useful intel


**3.2 TF-IDF (Term Frequency–Inverse Document Frequency)**

TF-IDF assigns importance to words based on:
1.   Frequency in document: How many times the word appears in that sentence/report.
2.   Rarity across documents: How rare the word is in all reports.


If a word appears in many reports → It is less important.
If it appears in few reports → It is more important.




```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "enemy activity detected",
    "enemy security forces deployed"
]

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(documents)

print("Features:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", X.toarray())
```
