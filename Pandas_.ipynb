3.5 .loc and .iloc (Row selection)

```python
df_ = pd.DataFrame(
    {
        'Product': ['Pen', 'Notebook', 'Pencil', 'Eraser'],
        'Price': [10, 50, 5, 8],
        'Quantity': [100, 40, 200, 150]
    },
    index=['P1', 'P2', 'P3', 'P4']
)

df_
```

```python
print(df_.loc['P1'])                 # Product with ID P1
print(df_.loc['P2':'P3'])            # Products P2 and P3
print(df_.loc[['P1', 'P4']])         # Selected products
print(df_.loc['P3', 'Price'])        # Price of P3
print(df_.loc[:, 'Quantity'])        # Quantity column
```

```python
print(df_.iloc[0])            # First product
print(df_.iloc[1:3])            # Products at pos 1 and 2
print(df_.iloc[[0, 3]])       # First and last products
print(df_.iloc[2, 1])         #  Price at pos (2,1)
print(df_.iloc[:, 1])         # Price column
```

.loc  → uses labels → end index included

.iloc → uses numbers → end index excluded


Use .loc when you know the name of the data, and .iloc when you know the position.

4. **Handling Missing Values (NaN / NA)**

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['A', 'B', 'C'],
    'Age': [22, np.nan, 25],
    'Salary': [30000, 40000, np.nan]
}

df_2= pd.DataFrame(data)
df_2
```

Identify missing values

```python
df_2.isna()  #Returns True where data is missing.
```

Count missing values column-wise

```python
df_2.isna().sum()
```

```python
df_2.dropna()    #Drop rows containing missing values
```

Fill Missing Values

```python
df_2['Age'].fillna(df_2['Age'].mean(), inplace=True)   #Fill Age with Mean
```

```python
df_2['Salary'].fillna(0, inplace=True)   #Fill Salary with 0
df_2
```

**5. GroupBy**

GroupBy is used to: Split data into groups, Apply a function (mean, sum, count), Combine results

```python
data = {
    'Dept': ['IT', 'IT', 'HR', 'HR'],
    'Salary': [50000, 60000, 45000, 48000]
}

df_emp= pd.DataFrame(data)
df_emp
```

```python
df_emp.groupby('Dept')['Salary'].mean()     #Average Salary per Department
```

```python
df_emp.groupby('Dept').agg(                          #Multiple Aggregations
    MinSalary=('Salary', 'min'),
    MaxSalary=('Salary', 'max'),
    AvgSalary=('Salary', 'mean')
)SQL query to a relational database
```

6. **Merge and Join**

In real-world data, information is stored in multiple tables.
merge() combines them like SQL JOIN.

```python
emp = pd.DataFrame({
    'EmpID': [1, 2, 3],
    'Name': ['Amit', 'Riya', 'John']
})

dept = pd.DataFrame({
    'EmpID': [1, 2, 3],
    'Department': ['IT', 'HR', 'Finance']
})
```

```python
pd.merge(emp, dept, on='EmpID')  #Merge on Common Column
```

```python
pd.merge(emp, dept, how='left', on='EmpID')  #Left Join
```

```python
pd.merge(emp, dept, how='outer', on='EmpID')   #outer join
```

7. **Time-Series Data in Pandas**

Data indexed by date or time (sales, stock prices, temperature).Pandas has built-in support for time-series analysis.

```python
dates = pd.date_range('2024-01-01', periods=5)

sales = pd.Series([100, 120, 90, 150, 200], index=dates)
sales
```

```python
sales['2024-01-02':'2024-01-04']    #Date-Based Selection
```

```python
sales.resample('M').sum()           #Converts daily data to monthly summary.
```

8. Window function

Window functions perform calculations on a moving window of data.

Used for: Moving averages, Trend analysis, Smoothing data

```python
sales.rolling(window=2).mean()   #Rolling Mean (Moving Average)
                                 #Each value is the average of the previous 2 values.
```

```python
sales.expanding().mean()          #Expanding Mean
                                  #Average from start until current row.
```

```python
sales.ewm(span=2).mean()           #Exponential Moving Average (EMA)
                                   #Gives more weight to recent values.
```

9. **Complete ETL Example Using Pandas**

9.1 Extract-Load Raw Data into DataFrame

```python
import pandas as pd
import numpy as np

data = {
    'EmpID': [101,102,103,104,105,106,107,108,109,110,111,112,113,114,115],
    'Name': ['Amit','Riya','John','Neha','Raj','Kiran','Anu','Vijay','Pooja','Sam','Arjun','Meena','Rohit','Sara','Dev'],
    'Department': ['IT','HR','IT','Finance','HR','IT','Finance','IT','HR','IT','Finance','HR','IT','Finance','HR'],
    'Age': [25, np.nan, 30, 28, 35, 26, np.nan, 29, 31, 27, 34, np.nan, 28, 32, 26],
    'Salary': [50000, 45000, np.nan, 55000, 48000, 52000, 51000, np.nan, 47000, 54000, 60000, 46000, np.nan, 58000, 49000],
    'JoiningDate': ['2022-01-10','2022-03-15','', '2021-07-01','2020-06-20',
                    '2022-02-18','2021-09-12','', '2023-01-05','2022-11-25',
                    '2020-04-10','', '2023-03-15','2021-12-01','2022-08-30']
}

emp_details= pd.DataFrame(data)
emp_details
```

9.2 Transform- Clean and Process the Data

Understand the Data

```python
emp_details.info()
```

Handle Missing Age Values

```python
emp_details['Age'].fillna(emp_details['Age'].mean(), inplace=True)
```

```python
emp_details['Salary'].fillna(emp_details['Salary'].median(), inplace=True)
```

Clean Joining Date Column

```python
emp_details['JoiningDate'].replace('', np.nan, inplace=True)
emp_details['JoiningDate'].fillna('2022-01-01', inplace=True)
emp_details['JoiningDate'] = pd.to_datetime(emp_details['JoiningDate'])
```

```python
emp_details.head()
```

Standardize Department Names

```python
emp_details['Department'] = emp_details['Department'].str.upper()
```

Create New Columns

```python
emp_details['AnnualSalary'] = emp_details['Salary'] * 12
```

```python
emp_details.head()
```

Rename Columns for Clarity

```python
emp_details.rename(columns={'JoiningDate': 'DOJ'}, inplace=True)
```

```python
emp_details.head()
```

9.3 Load- Save the Cleaned Data

```python
emp_details.to_csv("employees_cleaned.csv", index=False)
```

**SQL lite**

```python
engine = create_engine("sqlite:///company.db")
```

Reading SQL Data into Pandas

```python
df = pd.read_sql("SELECT * FROM employees", engine)             #Read Entire Table
df
```

```python
pd.read_sql(
    "SELECT Name, Salary FROM employees",                       #Select Specific Column
    engine
)
```

```python
pd.read_sql(
    "SELECT * FROM employees WHERE Dept = 'IT'",                #Filter rows
    engine
)
```

```python
pd.read_sql(
    "SELECT * FROM employees ORDER BY Salary DESC",            #sort result
    engine
)
```
