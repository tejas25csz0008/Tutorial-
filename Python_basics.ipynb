**1.5 Control Statements in Python**

In real operations, decisions depend on conditions:

If threat detected → Send alert

If weather bad → Delay mission

If fuel low → Return to base

Similarly, computers use control statements to decide what to do.

**Types of Control Statements**

1.If Statement → Single decision

2.If–Else → Two choices

3.If–Elif–Else → Multiple choices

4.Loops → Repeat tasks

5.For loop

6.While loop


**Examples**

```python
weather = "Clear"
```

```python
if weather == "Clear":
    print("Mission Go")
else:
    print("Mission Postponed")
```

```python
weapons = ["Rifle", "Drone", "Sniper"]

for weapon in weapons:
    print("Check:", weapon)
```

**Write a Python program that:**

**Creates:**

1.One integer variable for number of items

2.One float variable for price per item

3.One string variable for customer name

4.Create a list of 4 grocery items.

5.Create a function called calculate_total() that:

    Calculates total cost (total = items * price)

    Prints total bill

6.Use a control statement:

    If total bill is greater than 500 → Print “High Bill”

    Else → Print “Normal Bill”


```python
customer_name = "Tejas"   # string
number_of_items = 4       # integer
price_per_item = 150.50   # float

grocery_items = ["Rice", "Milk", "Bread", "Fruits"]


def calculate_total(items, price):

    total = items * price
    print("Customer:", customer_name)
    print("Total Bill:", total)

    return total


total_amount = calculate_total(number_of_items, price_per_item)


if total_amount > 500:
    print("High Bill")
else:
    print("Normal Bill")


```
