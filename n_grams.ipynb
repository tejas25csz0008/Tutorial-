Because “how to reset password” is common.

That’s a Trigram pattern.

**What is an N-Gram**




An N-gram is a group of N consecutive words.

Where:

N = Number of words together

Gram = Group

**Types of N-Grams**

| Type    | N       | Example             |
| ------- | ------- | ------------------- |
| Unigram | 1 word  | drone               |
| Bigram  | 2 words | drone detected      |
| Trigram | 3 words | drone detected near |

**Why Do we need increase in n value?**

Single words often do not give full meaning.

Example:

Word → border

This alone gives limited information.

But:

Bigram → border security

Now meaning becomes clearer.

Even better:

Trigram → border security alert

Now context is strong.

N-grams help computers understand word combinations and context.

Reports:

Enemy movement in sector

Enemy movement in sector

Enemy movement in area

| Phrase    | Count |
| --------- | ----- |
| in sector | 2     |
| in area   | 1     |


Now prediction:

Enemy movement in → sector (higher probability)

N-gram probability = The most frequently seen word comes next.

**Task**

Predict the most likely word after:

**Drone detected near ___ ?**

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter
import nltk

nltk.download('punkt')

reports = [
    "Drone detected near border",
    "Drone detected near base",
    "Drone detected near border",
    "Drone detected near camp"
]

# Combine all text
text = " ".join(reports)

tokens = word_tokenize(text)

# Create bigrams
bigrams = list(ngrams(tokens, 2))

# Count frequency
bigram_counts = Counter(bigrams)

# Filter bigrams starting with "near"
near_bigrams = {
    pair: count
    for pair, count in bigram_counts.items()
    if pair[0] == "near"
}

print("Bigram Frequencies:", near_bigrams)
```

**Exercise**

Using the text:

"Drone surveillance system detects drone movement"


Perform:

Generate Unigrams

Generate Bigrams

Generate Trigrams

Predict next word after:

“drone surveillance ___”

“detects drone ___”

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import nltk

# Download tokenizer (run once)
nltk.download('punkt')

# Input text
text = "Drone surveillance system detects drone movement"

# Tokenize
tokens = word_tokenize(text)

print("Tokens:", tokens)

# Generate N-grams
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("\nUnigrams:\n", unigrams)
print("\nBigrams:\n", bigrams)
print("\nTrigrams:\n", trigrams)
```
